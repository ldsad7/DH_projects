```python
import urllib.request
from urllib.parse import quote
import re
import csv
import os
import time
```

```python
langs = dict()
with open('extra.txt', 'r', encoding='utf-8') as f:
    for line in f.readlines():
        t = line.strip().split('        ')
        if ',' not in t[1]:
            langs[t[1]] = t[2].split(', ')
        else:
            langs[t[1].split(', ')[0]] = t[2].split(', ')
            langs[t[1].split(', ')[1]] = t[3].split(', ')
```

```python
langs_ = set(langs.keys())
```

```python
main = 'https://ru.wiktionary.org/wiki/'
words = dict()
for elem in langs_:
    url = main + quote(elem)
    req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/6.1 (Windows NT 6.1; Win64; x64)'})
    try:
        response = urllib.request.urlopen(req)
        html = response.read().decode('utf-8')
        a = re.findall('<td[^<>]*?>([^<>]+)<\/td>', html)
        for t in a:
            t = re.sub('а́', 'а', t)
            t = re.sub('е́', 'е', t)
            t = re.sub('и́', 'и', t)
            t = re.sub('о́', 'о', t)
            t = re.sub('у́', 'у', t)
            t = re.sub('э́', 'э', t)
            t = re.sub('ю́', 'ю', t)
            t = re.sub('я́', 'я', t)
            t = re.sub('ы́', 'ы', t)
            if elem in words:
                words[elem].append(t)
            else:
                words[elem] = [t]
    except:
        print(elem)
        pass
```

```python
for key, value in words.items():
    for i, elem in enumerate(value):
        words[key][i] = re.sub('ы́', 'ы', value[i])
```
