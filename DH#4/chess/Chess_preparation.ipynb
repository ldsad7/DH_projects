```python
for i in range(83942, 100000):
    print(i)
    href = 'http://www.chessebook.com/world_reti.php?lan=ru&t=op&pos_ID={}'.format(i)
    page = requests.get(href, headers=headers)
    html_ = html.unescape(page.text)
    title = re.findall('<TITLE>(.*?)<\/TITLE>', html_)[0].split()
    first = title[0].split(',')
    if len(first) < 2:
        print('!')
        continue
    first = first[1] + '. ' + first[0]
    second = title[2].split(',')
    if len(second) < 2:
        print('!')
        continue
    second = second[1] + '. ' + second[0]
    tmp = re.findall('<tr align="">(.*?)<\/tr>', html_)[0]
    arr = re.findall('<a.*?>(.*?)<.*?\/a>', tmp)
    arr = [elem.replace('x', '-') for elem in arr]
    if first in players:
        players[first].append([elem for i, elem in enumerate(arr) if i % 2 == 0])
    else:
        players[first] = [[elem for i, elem in enumerate(arr) if i % 2 == 0]]
    if second in players:
        players[second].append([elem for i, elem in enumerate(arr) if i % 2 == 1])
    else:
        players[second] = [[elem for i, elem in enumerate(arr) if i % 2 == 1]]
```
